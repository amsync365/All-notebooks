Q3: Which vaccines dominate globally, and which countries contribute the most to each vaccine?
</div>

```python

# aggregate total doses by vaccine and by (vaccine, country)
by_vaccine = df2.groupby('vaccine', as_index=False)['total_vaccinations'].sum()
by_vaccine_country = df2.groupby(['vaccine','location'], as_index=False)['total_vaccinations'].sum()

# optional: keep only top contributors per vaccine to avoid فوق‌العاده شلوغ شدن
# here we keep top 10 countries per vaccine (tweak n as you like)
n = 10
top_contributors = (
    by_vaccine_country
    .groupby('vaccine', group_keys=False)
    .apply(lambda g: g.nlargest(n, 'total_vaccinations'))
    .reset_index(drop=True)
)

# create a combined dataframe: top level = vaccine, second level = country contributions
# for sunburst we need each row to belong to a path; include the vaccine totals as rows too.
vaccine_rows = by_vaccine.rename(columns={'total_vaccinations':'doses'})
country_rows = top_contributors.rename(columns={'total_vaccinations':'doses'})

# add a helper column for root path (optional)
# Build a single dataframe with path ['vaccine','location']
sun_df = pd.concat([
    # vaccine-level rows: have vaccine but no location (location set to "")
    vaccine_rows.assign(location=""),
    # country-level rows: vaccine + location
    country_rows
], ignore_index=True)

# draw sunburst: first ring = vaccine (global share), second ring = country contributors
fig = px.sunburst(
    sun_df,
    path=['vaccine','location'],   # levels: vaccine -> country
    values='doses',
    title='Global Vaccine Market Share and Top Country Contributors',
    height=600
)
fig.update_traces(textinfo='label+percent entry')  # show percent of parent
fig.show()
```

    <ul style="margin:0;padding-left:20px;">
        <li>
            Based on data quality, reporting consistency, and vaccination coverage, <strong>Germany</strong> has been identified as the top country for building a predictive model of COVID-19 vaccination trends.
        </li>
    </ul>
</div>

```python
#Number of reporting days in the main dataset (per country)
days_main = df1.groupby('country')['date'].nunique()

#Build a daily table from manufacturer dataset (sum values if multiple records exist per day)
manu_daily = df2.groupby(['location', 'date'], as_index=False)['total_vaccinations'].sum()
manu_days = manu_daily.groupby('location')['date'].nunique()

#Merge on (country, date) and compute mean relative error (no loops)
merged = df1[['country','date','total_vaccinations']].merge(
    manu_daily.rename(columns={'location':'country','total_vaccinations':'total_manu'}),
    on=['country','date'],
    how='inner'
)

error_percent = (
    merged.assign(
        rel_error=lambda d: (d['total_vaccinations'] - d['total_manu']).abs()
                            / d['total_vaccinations'].replace(0, np.nan)
    )
    .groupby('country')['rel_error']
    .mean() * 100
)

#Combine indicators into a DataFrame and compute the final score
ranking = pd.DataFrame({
    'days_main': days_main,
    'days_manu': manu_days
}).reindex(days_main.index)

ranking['days_manu'] = ranking['days_manu'].fillna(0).astype(int)
ranking['error_percent'] = error_percent.reindex(ranking.index)

# Weighted scoring formula
ranking['score'] = (
    0.5 * (ranking['days_main'] / ranking['days_main'].max()) +  # main reporting completeness
    0.3 * (ranking['days_manu'] > 50).astype(int) +              # manufacturer coverage bonus
    0.2 * (1 - ranking['error_percent'].fillna(100) / 100)        # error penalty
)

final_ranking = ranking.sort_values('score', ascending=False).round(4)

display(
    final_ranking.head(15)[['days_main','days_manu','error_percent','score']]
    .sort_values(by='error_percent')
)
```
